## 转换时间类型

```python
train_data['deal_time']=pd.to_datetime(train_data['deal_time'],format='%Y%m%d%H')
```

### 将数据按线路切割

```python
line_281_data = train_data[train_data['line_name']==281]
line_281_data.head()
```

```python
line_565_data = train_data[train_data['line_name']==565]
line_565_data.head()
```

###  以小时为单位统计人次（已切割数据）

```python
line_565_groupbytime=pd.DataFrame(line_565_groupbytime).reset_index()
line_565_groupbytime=line_565_data.groupby('deal_time').count()['card_id']
line_565_groupbytime.head()
```

```python
line_281_groupbytime = line_281_data.groupby('deal_time').count()['card_id']
line_281_groupbytime = pd.DataFrame(line_281_groupbytime).reset_index()
line_281_groupbytime.head()
```
